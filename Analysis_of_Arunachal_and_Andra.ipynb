```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/election_results_Andhra_Pradesh.csv')
data2 = pd.read_csv('/content/election_results_Arunachal_Pradesh.csv')
data3 = pd.read_csv('/content/election_results_Odisha.csv')
data4 = pd.read_csv('/content/election_results_Sikkim.csv')
```

```python


from matplotlib import pyplot as plt
data1['Margin'].plot(kind='hist', bins=20, title='Margin')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


from matplotlib import pyplot as plt
data1['Const. No.'].plot(kind='hist', bins=20, title='Const. No.')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


from matplotlib import pyplot as plt
data1.plot(kind='scatter', x='Const. No.', y='Margin', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


from matplotlib import pyplot as plt
import seaborn as sns
data1.groupby('Leading Party').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


from matplotlib import pyplot as plt
import seaborn as sns
data1.groupby('Trailing Party').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Trailing Party'].value_counts()
    for x_label, grp in data1.groupby('Leading Party')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Leading Party')
_ = plt.ylabel('Trailing Party')
```
